```python
from fastai.vision import *
import seaborn as sns
from matplotlib.pyplot import figure
from fastai.callbacks import * 
```

```python
tfms = get_transforms(max_zoom=1.3,max_warp=0.4, max_lighting= 0.4)
```

```python
np.random.seed(369)
src = (ImageList.from_csv('/home/jupyter/tutorials/data/test925','filename.csv')
       .split_by_rand_pct(0.2)
       .label_from_df(label_delim=' '))
```

```python
data = (src.transform(tfms,size=64)
        .databunch().normalize(imagenet_stats))
```

```python
acc_02 = partial(accuracy_thresh, thresh=0.2)
f_score = partial(fbeta, thresh=0.2)
learn = cnn_learner(data, models.resnet34, metrics=[acc_02, f_score])
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10,3e-02,callbacks=[ShowGraph(learn),SaveModelCallback(learn)])
```

```python
learn.recorder.plot_losses()
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(20,slice(1e-05,9e-05),callbacks=[ShowGraph(learn),SaveModelCallback(learn)])
```

```python
learn.recorder.plot_losses()
```
